```python
import pandas as pd
from sklearn.datasets import load_iris
```

## 載入資料 ＆ 資料前處理
這dataset很乾淨所以只要把資料載入，並做成DataFrame就好

```python
iris= load_iris()
df=pd.DataFrame(iris['data'],columns=iris['feature_names'],)
df['Ans']=iris['target']
```

```python
iris.keys()
```

## import  train_test_split 
從dataset中分割資料，切一部分出來做驗證資料集

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris['data'],iris['target'],test_size=0.1)
```

## 畫圖

```python
import seaborn as sns
sns.set()
i=sns.load_dataset("iris")

g = sns.lmplot(x="sepal length (cm)", y="sepal width (cm)", hue="Ans",
               height=5, data=df)
g.set_axis_labels("Sepal length (mm)", "Sepal width (mm)")
```

# import 決策術分類方法，並fit training data

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(max_depth=2)
clf.fit(x_train,y_train)
```

## 使用accuracy_score比較 預測結果與真正結果

```python
pre=clf.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pre)
```

# graphviz 畫圖

```python
from sklearn.tree import export_graphviz
import graphviz
g=export_graphviz(clf,feature_names=iris['feature_names'],class_names=iris['target_names'],filled=True)
graphviz.Source(g)
```
